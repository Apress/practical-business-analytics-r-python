# Pre-processing data using APPLY()

```python
def pre_proc(x):
    x = x.replace('$',"")
    x = x.replace('[&]',"")
    x = x.strip()
    x = x.replace('[/]',"")
    x = x.lower()
    return(x)
```

```python
products
```

```python
products.dtypes
```

```python
## Some values in Unit price column has $ sign and some do not. To keep the values consistent
## we will remove $ sign from the values.
## We will call pre_proc() function
```

```python
products['Unit Price'] = products['Unit Price'].apply(pre_proc)
```

```python
products
```

```python
products
```

```python
products['Product Name'] = products['Product Name'].str.replace('-'," ")
```

```python
products
```

```python
products['Product Name'] = products['Product Name'].str.upper()
```

```python
products
```

```python
products['Product Name'] = products['Product Name'].str.capitalize()
```
