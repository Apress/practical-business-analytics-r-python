```python
import os
import pandas as pd
import numpy as nd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data_dir = 'C:/Personal/dataset'
os.chdir(data_dir)
```

```python
marys_df = pd.read_csv('marys.csv')
```

```python
marys_df.head()
```

```python
print(marys_df.shape)
```

```python
marys_df2=marys_df.drop(columns = {'Trans. Id'})
```

```python
marys_df2.head()
```

```python
item_sets = apriori(marys_df2, min_support=0.5, use_colnames=True, low_memory=False, verbose=1)
```

```python
item_sets.head(6)
```

```python
mba_rules = association_rules(item_sets, 
                              metric="confidence", 
                              min_threshold=0.5,
                             support_only=False).sort_values(by = ['lift', 'confidence', 'support'], ascending =False)
```

```python
mba_rules.head(10)
```

```python
mba_rules[mba_rules['antecedents']==frozenset({'Shirt'})].sort_values(by = 'lift', ascending = False)
```

```python
mba_rules[mba_rules['antecedents']==frozenset({'Shirt','Wallet'})].sort_values(by = 'lift', ascending = False)
```
