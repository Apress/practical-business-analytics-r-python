```python
stocks=pd.read_csv("E:/Umesh-MAY2022/Personal-May2022/BA2ndEdition/2ndEdition/Book Chapters/Chapter 5 - EDA/stocks3.csv")
```

```python
#Histograms
```

```python
stocks.Stock1.plot.hist(by=None, bins=15)
```

```python
stocks.Stock1.plot.kde(bw_method='silverman', ind=1200)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(stocks.Stock1, kde=True, bins=15)
plt.show()
```

```python
#Scatter plots
```

```python
stocks.plot.scatter('Stock1', 'Stock2')
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(stocks,alpha=0.5, figsize=(10,6), ax=None, grid=False,
              color='red', hist_kwds={'bins':10, 'color':'red'})
```

```python
##BOX PLOT
```

```python
stocks.Stock1.plot.box()
```

```python
stocks.boxplot(column=['Stock1','Stock2','Stock3','Stock5'])
```

```python
import matplotlib.pyplot as plt
corel=stocks.corr()
corel.style.background_gradient(cmap='coolwarm').set_precision(3)
```

```python
plt.matshow(stocks.corr(method='pearson'))
plt.show()
```

```python
stocks1=stocks.drop(columns=['Day'])
```

```python
stocks1.plot.density(figsize = (10, 6),
                       linewidth = 1)
```

```python
stocks.Ratings.value_counts().plot.bar()
```
